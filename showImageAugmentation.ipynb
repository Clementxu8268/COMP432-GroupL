```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import torchvision.transforms.functional as F
from torchvision import transforms

transform = transforms.Compose([

    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    # transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

# sample_image_path = './ColorectalCancer/MUS/MUS-AADNSWYC.tif'  
sample_image_path = './Dataset1/ColorectalCancer/MUS/MUS-AADGLEEQ.tif'  

original_image = Image.open(sample_image_path)

plt.figure(figsize=(10, 5))
plt.imshow(original_image)
plt.title("Dataset 1 Colorectal Cancer")
plt.axis('off')
plt.show()

transformed_image = transform(original_image)  
transformed_image = F.to_pil_image(transformed_image) 

# 
plt.figure()
plt.imshow(transformed_image)
plt.axis('off')
plt.show()
```
